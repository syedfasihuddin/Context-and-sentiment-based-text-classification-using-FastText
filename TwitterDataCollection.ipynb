```python
tweets = api.user_timeline(screen_name = 'BarackObama', count = 100)
files_of_tweets = 'BarackObama.json'
with open(files_of_tweets, 'w') as outfile:
    for tweet in tweets:
        json.dump(tweet._json, outfile, sort_keys=True)
        outfile.write('\n')
        print(tweet.text)
```
